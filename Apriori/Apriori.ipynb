```python
# Import necessary libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Load the transaction data
# Read the CSV file without headers
transaction_data = pd.read_csv('apriori_transaction_data.csv', header=None)

# One-hot encode the DataFrame
# Using pd.get_dummies on a stacked version of the DataFrame
onehot = transaction_data.stack().str.get_dummies().reset_index(level=1, drop=True).groupby(level=0).sum()

print(onehot)

# Step 3: Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(onehot, min_support=0.01, use_colnames=True)

# Step 4: Generate the association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)

# Display the frequent itemsets and the generated rules
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```
